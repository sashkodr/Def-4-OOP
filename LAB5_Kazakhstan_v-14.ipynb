3. Підрахунок частоти символів у рядку:

```python
def count_symbol_frequency(text):
    """
    Функція для варіанту 14.
    Підраховує частоту появи кожного символу у введеному рядку.
    Повертає словник, де ключ — символ, а значення — кількість його входжень.
    """

    # Контроль типу введених даних
    if not isinstance(text, str):
        raise TypeError("Потрібно ввести саме рядок, а не число чи інший тип даних.")

    # Контроль порожнього рядка
    if len(text.strip()) == 0:
        raise ValueError("Рядок не може бути порожнім.")

    frequency = {}

    for char in text:
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1

    return frequency
```

4. **Результати тестування функції та контроль правильності введення даних**

```python
print("--- Тестування функції count_symbol_frequency ---")

test_1 = "Казахстан"
test_2 = "Astana"
test_3 = "Google Colab"

print(f"Рядок: {test_1}")
print(count_symbol_frequency(test_1))
print()

print(f"Рядок: {test_2}")
print(count_symbol_frequency(test_2))
print()

print(f"Рядок: {test_3}")
print(count_symbol_frequency(test_3))
print()

# Контроль неправильного введення даних
print("--- Контроль введення ---")

try:
    print(count_symbol_frequency(12345))
except Exception as e:
    print("Помилка:", e)

try:
    print(count_symbol_frequency("   "))
except Exception as e:
    print("Помилка:", e)
```

5. **Графічна візуалізація результатів роботи:

```python
import matplotlib.pyplot as plt

# Рядок для візуалізації
def visualize_symbol_frequency(text):
    freq = count_symbol_frequency(text)

    symbols = list(freq.keys())
    counts = list(freq.values())

    plt.figure(figsize=(9, 5))
    plt.bar(symbols, counts)
    plt.title(f"Частота символів у рядку: {text}")
    plt.xlabel("Символи")
    plt.ylabel("Кількість повторень")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

visualize_symbol_frequency("KAZAKHSTAN")
```
